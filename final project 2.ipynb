#### 4. Logistic Regression Modeling

Logistic regression is a statistical modeling technique used to analyze the relationship between one or more predictor variables and a binary outcome variable.Since our outcome variable Attrition is a binary variable, we plan to use logistic regression in this analysis.

We followed this formula:

$$
\log\frac{p}{1-p} = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p
$$

where p is the success probability, the probability of the binary outcome variable being a "Yes" or "No" outcome, given the values of the predictor variables. 

$ X_1 $ = Job Satisfaction (the variable of interest)

$ X_2 $ ... $ X_p $ = control variables

#### 5. Terms of Predictors

Since we would use logistic regression in this analysis, we do not need to follow the assumption that predictor variable follows a normal distribution. Therefore, we would not use logarithm terms in our model.

Also, using higher-order terms on our categorical predictor variable Job Satisfaction when predicting a binary outcome is not be meaningful. A categorical variable has a limited number of possible values, and using higher-order terms may not provide any additional information about the relationship between the predictor variable and the outcome variable. Therefore, we would not use logarithm terms in our model.

#### 6. Two Stage Residual Inclusion
Adding instrumental variables is a potential way to make our estimation unbiased. We planned to try Two Stage Residual Inclusion (2SRI) to conduct the regression as well.

In 2SRI, the first stage involves regressing the potentially endogenous predictor variable (Job Satisfaction) on a set of instrumental variables that are correlated with the endogenous variable but not with the error term. The resulting residuals from this first stage regression are then included as additional predictor variables in the main regression model.

The second stage of 2SRI involves estimating the coefficients of the main regression model using the residuals obtained from the first stage as additional predictor variables. This approach helps to eliminate the endogeneity bias by using the instrumental variables to obtain unbiased estimates of the endogenous predictors and then adjusting for the remaining unobserved confounding factors using the residuals.

Note: In order to obtain effective residuals in the first stage, we used OLS regression. However, since we used categorical variables for analysis, we were unable to verify the normal distribution of the data to demonstrate that it meets the assumptions of OLS. This is a problem that needs to be continuously improved in this project.

## Modeling

> We first conducted a logistic regression with Attrition as the outcome variable and Job Satisfaction as the only predictor.

```python
y = df["Attrition"]
X = df["JobSatisfaction"]

X = sm.add_constant(X)
model = sm.Logit(y, X)
result = model.fit()
print(result.summary())
```

According the model summary, we could interpretate that if the job satisfaction of an employee increases by 1, log odds of success decreases by 0.25. More specifically, the odds of attrition (i.e., the ratio of the probability of attrition to the probability of no attrition) decrease by a factor of e^(-0.25) ≈ 0.78 for each unit increase in job satisfaction. There is a negative relationship between job satisfaction and the likelihood of attrition. 

However, the R squared of this model is only 0.01, which is low. The model does not fitt the data well. There may be some omitted variables that will confound our estimation.

> We added the control variables into the model and conduct logitic regression again.

```python
# control variables added
X1 = df[["JobSatisfaction","Age", "Gender", "Education", "TotalWorkingYears","MonthlyIncome","PerformanceRating","WorkLifeBalance",
         "NumCompaniesWorked","EnvironmentSatisfaction","YearsAtCompany","YearsInCurrentRole","YearsWithCurrManager","MaritalStatus_Divorced",
         "MaritalStatus_Married","MaritalStatus_Single","RelationshipSatisfaction","PercentSalaryHike","YearsSinceLastPromotion"]]

X1 = sm.add_constant(X1)

# create a logistic regression model
model1 = sm.Logit(y, X1)

# fit the model to the data
result1 = model1.fit()

# print the summary statistics of the model
print(result1.summary())
```

Compared to the first model we ran using only Job Satisfaction as the predictor, the model with control variables had higher R-squared(0.15).

> To further improve our regression result, we added instrumental variables in this analysis using 2 Stage Residual Inclusion.

```python
import statsmodels.formula.api as smf
df.rename(columns={'BusinessTravel_Non-Travel': 'BusinessTravel_Non_Travel'}, inplace=True)

reg_ivrjob = smf.ols("JobSatisfaction ~ DistanceFromHome +  BusinessTravel_Non_Travel + BusinessTravel_Travel_Frequently + BusinessTravel_Travel_Rarely + Age + Gender + Education + TotalWorkingYears + MonthlyIncome + PerformanceRating + WorkLifeBalance + EnvironmentSatisfaction  + NumCompaniesWorked + YearsAtCompany + YearsInCurrentRole + YearsWithCurrManager + MaritalStatus_Divorced + MaritalStatus_Married + MaritalStatus_Single + RelationshipSatisfaction + PercentSalaryHike +  YearsSinceLastPromotion", data=df)
res_ivrjob = reg_ivrjob.fit()

df['jobsatis_resid'] = res_ivrjob.resid

reg_ivr = smf.logit('Attrition ~ Age + Gender + Education + TotalWorkingYears + MonthlyIncome + PerformanceRating + WorkLifeBalance + NumCompaniesWorked + EnvironmentSatisfaction  + YearsAtCompany + YearsInCurrentRole + YearsWithCurrManager + MaritalStatus_Divorced + MaritalStatus_Married + MaritalStatus_Single + RelationshipSatisfaction + PercentSalaryHike +  YearsSinceLastPromotion + jobsatis_resid', data=df) 
res_ivr = reg_ivr.fit()
print(res_ivr.summary())
```

After taking the instrumental variables into consideration, the R squared of the model slightly increased, indicating that the model is a better fit for the data, meaning that it can explain more of the variance in the dependent variable.

## Test Assumption

#### Assumption 1: Logistic regression requires the dependent variable to be binary or ordinal. 
The dependent variable, Attrition, we used in this analysis is binary.

#### Assumption 2: Logistic regression requires the observations to be independent of each other. In other words, the observations should not come from repeated measurements or matched data.

According to our data description, most of our variables are independent except for the variables related to income level. Among the daily rate, daily income, monthly rate, monthly income observations, we chose to use monthly income as our control variables since we believe they have carried repeated information about an employee’s salary level. 

#### Assumption 3: Logistic regression requires there to be little or no multicollinearity among the independent variables. This means that the independent variables should not be too highly correlated with each other. 

We have examine the corrleation between all independent variables in the heatmaps we shown above. Although some of the control variables have relatively high mcoolreation with each other, but this would not affect the estimation of Job Satisfaction, which is the our variable in interst.

#### Assumption 4: Logistic regression assumes linearity of independent variables and log odds. Although this analysis does not require the dependent and independent variables to be related linearly, it requires that the independent variables are linearly related to the log odds.

```python
X2 = df[["Age", "Gender", "Education", "TotalWorkingYears","MonthlyIncome","PerformanceRating","WorkLifeBalance",
         "NumCompaniesWorked","EnvironmentSatisfaction","YearsAtCompany","YearsInCurrentRole","YearsWithCurrManager","MaritalStatus_Divorced",
         "MaritalStatus_Married","MaritalStatus_Single", "RelationshipSatisfaction","PercentSalaryHike","YearsSinceLastPromotion","jobsatis_resid"]]

probs = res_ivr.predict(X2)

# Calculate log odds
log_odds = np.log(probs / (1 - probs))

# Create scatter plots
fig, axs = plt.subplots(4, 5, figsize=(20, 16))
for i, col in enumerate(X2.columns):
    axs[i//5, i%5].scatter(X2[col], log_odds)
    axs[i//5, i%5].set_xlabel(col)
    axs[i//5, i%5].set_ylabel('Log Odds')
plt.tight_layout()
plt.show()
```

The assumption of linearity in logistic regression refers to the linear relationship between the log odds of the dependent variable and the continuous independent variables.The above graphs are the scatter plots of the relationship between the independent variables and the log odds of the binary outcome, Attrition.

#### Assumption 5: Logistic regression typically requires a large sample size. A general guideline is that we need at minimum of 10 cases with the least frequent outcome for each independent variable in your model. 

Counting all the independent variables we have, including the 1 variable of interest (Job Satisfaction), 18 control variables, and 4 instrumental variables, we have 23 independent variables in toall. The probability of least frequency outcome is 0.1612. 

$$ \text{ probability of least frequency outcome } = \frac{ \text{count of attrition} }{(\text{count of stay + count of attrtion}) }$$

$$ 0.1612 = 237/1470 $$

Therefore, the minimum sample size of our dataset should be 10*23/0.1612 = 1427. Since our dataset has 1470 observations, it fulfills this assumption.  

## Improve the Model 

#### 1. Instrumental Variables
We can make improvements in the selection of instrumental variables. Instrumental variables should be factors that affect our independent variable, which is job satisfaction, but do not have an effect on the dependent variable, which is attrition. Such variables include the weather when employees filled out the job satisfaction survey. Weather can affect employees' moods and make them more likely to express dissatisfaction with their work by giving lower scores. Monday may not be a good time for employees to provide objective responses in the survey, as people tend to have negative emotions on the first day of work after the weekend. Another factor can be perceived anonymity. If employees feel that their survey responses are not truly anonymous, they may be less likely to give honest feedback, leading to an inaccurate representation of their true level of job satisfaction. Interaction terms can help capture the non-linear relationship between the categorical variable and the dependent variable. Interaction terms are created by multiplying the categorical variable with another independent variable in the model.

$$
